```python
import os
import pickle
import torch
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.spatial.distance import cdist
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from model import GaitModel

# =====================================
# CONFIGURATION
# =====================================

DATA_PATH = "data/processed"
WEIGHTS_PATH = "weights/gait_model_best.pth"

# Use only first 10 subjects for clean confusion matrix
TEST_SUBJECTS = [f"{i:03d}" for i in range(75, 85)]

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =====================================
# LOAD MODEL
# =====================================

model = GaitModel(num_classes=74).to(DEVICE)

checkpoint = torch.load(
    WEIGHTS_PATH,
    map_location=DEVICE
)

model.load_state_dict(
    checkpoint["model_state_dict"]
)

model.eval()

print("Best model loaded successfully.")

# =====================================
# FEATURE EXTRACTION FUNCTION
# =====================================

def get_feature(subject, condition, angle="090"):
    path = os.path.join(
        DATA_PATH,
        subject,
        condition,
        f"{angle}.pkl"
    )

    if not os.path.exists(path):
        return None

    with open(path, "rb") as f:
        sequence = pickle.load(f)

    if len(sequence) >= 30:
        sequence = sequence[:30]
    else:
        padding = np.tile(
            sequence[-1:],
            (30 - len(sequence), 1, 1)
        )
        sequence = np.concatenate(
            [sequence, padding],
            axis=0
        )

    inp = (
        torch.from_numpy(sequence)
        .float()
        .unsqueeze(0)
        .unsqueeze(2) / 255.0
    ).to(DEVICE)

    with torch.no_grad():
        embedding = model(inp)

    return embedding.squeeze().cpu().numpy()

# =====================================
# BUILD GALLERY
# Gallery = NM-01
# Probe = NM-02
# =====================================

gallery_features = []
gallery_labels = []

for subject in TEST_SUBJECTS:
    feat = get_feature(subject, "nm-01")

    if feat is not None:
        gallery_features.append(feat)
        gallery_labels.append(subject)

gallery_features = np.array(gallery_features)

# =====================================
# PREDICTIONS
# =====================================

true_labels = []
pred_labels = []

for subject in tqdm(TEST_SUBJECTS):

    probe_feat = get_feature(subject, "nm-02")

    if probe_feat is None:
        continue

    distances = cdist(
        probe_feat.reshape(1, -1),
        gallery_features,
        metric="cosine"
    )[0]

    pred_idx = np.argmin(distances)
    predicted_subject = gallery_labels[pred_idx]

    true_labels.append(subject)
    pred_labels.append(predicted_subject)

# =====================================
# CONFUSION MATRIX
# =====================================

cm = confusion_matrix(
    true_labels,
    pred_labels,
    labels=TEST_SUBJECTS
)

plt.figure(figsize=(10, 8))

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=TEST_SUBJECTS
)

disp.plot(
    cmap="Blues",
    xticks_rotation=45
)

plt.title("Confusion Matrix (First 10 Test Subjects)")
plt.show()
```
